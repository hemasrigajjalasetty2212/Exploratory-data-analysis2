```python
# IsolationForest
from sklearn.ensemble import IsolationForest

data =pd.read_csv("adult_with_headers.csv")
data_encoded=pd.get_dummies(data)
data_encoded
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(data_encoded)
```

```python
# predictions
y_pred = clf.predict(data_encoded)

#-1 for outliers and 1 for inliers.
y_pred
```

```python
data_encoded.loc[y_pred==-1]
```

```python
# Let us add a new data point which is outlier
# Get the number of columns in data_encoded
num_cols = len(data_encoded.columns)

# Create a list with num_cols values, filled with 0
new_row = [0] * num_cols

# Assign the desired values to the first 11 elements of new_row
new_row[:11] = [150,130,120,110,20,40,30,50,1,0,0]

# Assign the new_row to the DataFrame
data_encoded.loc[327] = new_row
data_encoded
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(data_encoded)
# predictions
y_pred_outliers = clf.predict(data_encoded)
y_pred_outliers
```

```python
# Calculate the anomaly scores
anomaly_scores = clf.decision_function(data_encoded)

# Add the anomaly scores to the DataFrame
data_encoded['scores'] = anomaly_scores
```

```python
# Remove the 'scores' column before prediction
data_encoded['anomaly'] = clf.predict(data_encoded.drop(columns=['scores']))
```

```python
#Print the outlier data points
data_encoded[data_encoded['anomaly']==-1]
```
